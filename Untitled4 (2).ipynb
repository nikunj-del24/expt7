```python
# write a python program to find factorial of a number.
n=int(input("enter the number:"))
i=1
while n>0:
  i*=n
  n-=1
print(i)
```

```python
# write a python program to find fibonacci series for n terms.
n=int(input("enter the number:"))
a=0
b=1
i=1
while i<=n:
  print(a,end=" ")
  c=a+b
  a=b
  b=c
  i+=1
```

```python
# write a python program to find fibonacci series by limiting it.
limit = int(input("Enter the limit:"))
a=0
b=1
while a<= limit:
  print(a,end=" ")
  a,b=b,a+b
```

```python
#reverse of a number
n=int(input("enter the number:"))
rev=0
while n>0:
  d=n % 10
  rev=rev * 10 + d
  n//=10
print("reverse number=",rev)

```

```python
#check palindrome number/string
n=int(input("Enter a number:"))
temp=n
rev=0
while n>0:
  rev=rev*10 + n%10
  n//=10
if temp==rev:
  print("Palindrome")
else:
  print("not palindrome")
```

```python
ch="madam"
i, j = 0,len(ch)-1
is_palindrome = True
while i<j:
  if ch[i]!=ch[j]:
    is_palindrome = False
    break
  i +=1
  j -=1
if is_palindrome:
  print("yes")
else:
  print("no")
```

```python
#check palindrome number
s=input("enter the string:")
rev=s[::-1]
if s==rev:
  print("palindrome")
else:
  print("not palindrome")
```

```python
#count digits in a number
n=int(input("enter the digit"))
count=0
while n>0:
  n//=10
  count+=1
print("Number of digits=",count)
```

```python
# search an element in a list
n=[10,20,30,40,50]
key=int(input("enter element to be searched"))
i=0
while i < len(n):
  if n[i] == key:
    print("element found at index", i)
    break
  i+=1
else:
  print("element not found")
```
